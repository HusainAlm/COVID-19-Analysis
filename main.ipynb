```python
sns.scatterplot(data=saudi_df, x='male_smokers', y='total_deaths', label='Male Smokers', color='blue')
sns.scatterplot(data=saudi_df, x='female_smokers', y='total_deaths', label='Female Smokers', color='red')

plt.xlabel('Smoking Rate (%)')
plt.ylabel('Total Deaths')
plt.title('Smoking vs COVID-19 Deaths in Saudi Arabia')
plt.legend()
plt.show()
```

```python
gcc_countries = ['Saudi Arabia', 'United Arab Emirates', 'Qatar', 'Kuwait', 'Oman', 'Bahrain']
gcc_df = world_df[world_df['location'].isin(gcc_countries)]
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=gcc_df, x='location', y='total_cases', estimator=max, ci=None)
plt.title('Total COVID-19 Cases in Gulf Countries')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12,6))
sns.barplot(data=gcc_df, x='location', y='total_deaths', estimator=max, ci=None, color='red')
plt.title('Total COVID-19 Deaths in Gulf Countries')
plt.xticks(rotation=45)
plt.show()
```

```python
gcc_df['death_rate'] = (gcc_df['total_deaths'] / gcc_df['population']) * 1e6

plt.figure(figsize=(12,6))
sns.barplot(data=gcc_df, x='location', y='death_rate', estimator=max, ci=None, color='darkred')
plt.title('COVID-19 Death Rate per Million in Gulf Countries')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=gcc_df, x='location', y='total_tests', estimator=max, ci=None, color='green')
plt.title('Total COVID-19 Tests in Gulf Countries')
plt.xticks(rotation=45)
plt.show()
```

```python
vaccine_df = pd.read_csv("C:/Users/zzyyx/Desktop/covid-19/Saudi_vaccine.csv")
```

```python
vaccine_df.head(5)
```

```python
print(vaccine_df[['people_vaccinated', 'people_fully_vaccinated', 'total_boosters']].isna().sum())
```

```python
vaccine_df[['people_vaccinated', 'people_fully_vaccinated', 'total_boosters']] = vaccine_df[['people_vaccinated', 'people_fully_vaccinated', 'total_boosters']].fillna(method='ffill')

stats = vaccine_df[['people_vaccinated', 'people_fully_vaccinated', 'total_boosters']].describe()
print(stats)
```

```python
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter, AutoDateLocator

vaccine_df['date'] = pd.to_datetime(vaccine_df['date'])

# Plot trends
plt.figure(figsize=(12, 6))
plt.plot(vaccine_df['date'], vaccine_df['people_vaccinated'], label='People Vaccinated', color='blue')
plt.plot(vaccine_df['date'], vaccine_df['people_fully_vaccinated'], label='People Fully Vaccinated', color='green')
plt.plot(vaccine_df['date'], vaccine_df['total_boosters'], label='Total Boosters', color='red')

plt.gca().xaxis.set_major_locator(AutoDateLocator())  # Automatically adjust date ticks
plt.gca().xaxis.set_major_formatter(DateFormatter('%b %Y'))  # Format dates as "Month Year"


plt.xlabel('Date')
plt.ylabel('Count')
plt.title('Vaccination Trends in Saudi Arabia')
plt.legend()
plt.grid(True)


plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

```

```python
vaccine_df['vaccination_coverage'] = (vaccine_df['people_fully_vaccinated'] / vaccine_df['people_vaccinated']) * 100
vaccine_df['booster_uptake'] = (vaccine_df['total_boosters'] / vaccine_df['people_fully_vaccinated']) * 100

# Display statistics
print(vaccine_df[['date', 'vaccination_coverage', 'booster_uptake']].dropna())
```
